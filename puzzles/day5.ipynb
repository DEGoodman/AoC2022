```python
import os

# get current location to ease relative file retrieval
NOTEBOOK_PATH = os.path.abspath("day5.ipynb")
# FILEPATH = os.path.join(os.path.dirname(NOTEBOOK_PATH), "../static/5/sample.txt")
FILEPATH = os.path.join(os.path.dirname(NOTEBOOK_PATH), "../static/5/input.txt")

input = []

# Read in file
with open(FILEPATH) as file:
    # collect the data
    input = file.read().splitlines()

# ok this one is going to be gross and hacky because unstructured text input is a nightmare 
# and I can't be bothered.

# like on day 2, I'm using the same gross feature flag to distinguish between part 1 and 2 logic
p1 = False

# Get the number of stacks
# again, because I'm lazy, I'm just going to grab this from the input text
# STACKS = 9; sample.txt
STACKS = 9

# Split the input by number of lines that relate to crates, intermediate lines we don't care about,
# and "directions" lines
# in the sample imput, creates are the first three lines
# INITIALSTACKHEIGHT = 3; sample.txt
INITIALSTACKHEIGHT = 8


# hardcoding the number of stacks I need because, once again, I'm lazy
crateStacks = [[],[],[],[],[],[],[],[],[]]

# in order to build our stacks properly, we must push them into the right stack.
# That means iterating from the bottom of our crate pile first
for i in range(INITIALSTACKHEIGHT):
    row = input[INITIALSTACKHEIGHT - i - 1]   
    
    # we can save a step by pushing these onto stacks at the same time as parsing the row
    stack = 0
    
    # current crate level parsed from line
    # each crate on input takes up 3 or 4 characters
    # the actual name of the crate is in position 2
    # for j in range(1,10, 4):; sample.txt
    for j in range(1, 36, 4):
        if row[j].isalpha():
            crateStacks[stack].append(row[j])
        stack += 1

    
# and then 2 lines of junk, so we skip them and go straight to the instructions
# for row in input[5:]:; sample.txt
for row in input[10:]:
    instructions = row.split(" ")
    # do what the instructions tell us! 
    # note: we can skip over the words, we don't need them
    # ['move', '1', 'from', '2', 'to', '1']
    # how many moves?
    
    # NOTE: part one moves the stack differently from part two. flagging this is easy and quick
    if p1:
        for moves in range(int(instructions[1])):
            # from where? adjust for off-by-1 err on stack location
            crate = crateStacks[int(instructions[3]) - 1].pop()
            # to where?adjust for off-by-1 err on stack location
            crateStacks[int(instructions[5]) - 1].append(crate)
    else: # part two
        # from stack m, get n crates
        # let's add some variables, because these are aweful to look at
        nCrates = int(instructions[1])
        fromStack = int(instructions[3]) - 1
        toStack = int(instructions[5]) - 1
        # do the stuff
        crates = crateStacks[fromStack][-nCrates:]
        # remove from original stack
        crateStacks[fromStack] = crateStacks[fromStack][:-nCrates]
        # append to target stack
        crateStacks[toStack] = crateStacks[toStack] + crates
    

# great! And now let's make a string of the top crates from each stack
crateList = [c.pop() for c in crateStacks]
topCrates = ''.join(crateList)
print("top crates: " + topCrates)
```
