```python
# hints:
#   "The first column is what your opponent is going to play: 
#    A for Rock, B for Paper, and C for Scissors. The second column--"
# ---
#   The second column, you reason, must be what you should play in response: 
#      X for Rock, Y for Paper, and Z for Scissors. Winning every time would be suspicious, 
#      so the responses must have been carefully chosen.
#
# Your total score is the sum of your scores for each round. 
# The score for a single round is the score for the shape you selected 
#    (1 for Rock, 2 for Paper, and 3 for Scissors) 
# plus the score for the outcome of the round 
#    (0 if you lost, 3 if the round was a draw, and 6 if you won).

import os

# get current location to ease relative file retrieval
NOTEBOOK_PATH = os.path.abspath("day2.ipynb")
# FILEPATH = os.path.join(os.path.dirname(NOTEBOOK_PATH), "../static/2/sample.txt")
FILEPATH = os.path.join(os.path.dirname(NOTEBOOK_PATH), "../static/2/input.txt")

# constants
SHAPE_SCORE = {"ROCK": 1, "PAPER": 2, "SCISSORS": 3}
SHAPE_MAP = {"A": "ROCK", "B": "PAPER", "C": "SCISSORS"}


total_points = 0
matches = []

# Read in file
with open(FILEPATH) as file:
    # check each line
    while (line := file.readline().rstrip()):
        matches.append(line.split(" "))


## -- Part 1 is significantly different from Part 2 , flagging this all out --##

p1 = False

# Puzzle 1
# What would your total score be if everything goes exactly according to your strategy guide?

def partOne():
    SELF_MAP = {"X": "ROCK", "Y": "PAPER", "Z": "SCISSORS"}

    # Flow:
    # 1. read each pairing
    # 2. calculate shape value
    # 3. determine pairing result
    # 4. calculate subtotal for match and increment total

    total = 0
    
    # Rock defeats Scissors, Scissors defeats Paper, Paper defeats Rock
    def matchResult(opponent, self):
        # tie
        if opponent == self:
            return 3
        # I win
        else:
            if (self == "ROCK" and opponent == "SCISSORS") or \
               (self == "SCISSORS" and opponent == "PAPER") or \
               (self == "PAPER" and opponent == "ROCK"):
                return 6
        # I lose
        return 0

    for pair in matches:
        # reset subtotal for each match-up
        subtotal = 0

        # increment shape value
        subtotal += SHAPE_SCORE[SELF_MAP[pair[1]]]
        print("shape subtotal: " + str(subtotal))

        # evaluate matchup
        match_score = matchResult(SHAPE_MAP[pair[0]], SELF_MAP[pair[1]])
        subtotal += match_score
        print("subtotal after pair match: " + str(subtotal))

        # increment total score
        total += subtotal
        print("total: " + str(total))

    return total

# Part 2
# "Anyway, the second column says how the round needs to end: 
#   X means you need to lose, 
#   Y means you need to end the round in a draw, and
#   Z means you need to win. Good luck!"

def partTwo():
    SCORE_GUIDE = {"X": 0, "Y": 3, "Z": 6}
    total = 0
    
    # Rock defeats Scissors, Scissors defeats Paper, Paper defeats Rock
    def shapeFinder(shape, outcome):
        # fuck it I'm just going to create a lookup table here
        possibilities = {
            ("ROCK", 0): "SCISSORS",
            ("ROCK", 6): "PAPER",
            ("PAPER", 0): "ROCK",
            ("PAPER", 6): "SCISSORS",
            ("SCISSORS", 0): "PAPER",
            ("SCISSORS", 6): "ROCK",
        }
        return possibilities[(shape, outcome)]
    
    for pair in matches:
        print("pair: " + str(pair))
        # reset subtotal for each match-up
        subtotal = 0
        
        # get opponent shape
        oppShape = SHAPE_MAP[pair[0]]
        print("opp shape: " + oppShape)
        
        # collect win/draw/lose value
        goal = SCORE_GUIDE[pair[1]]
        print("goal: " + str(goal))
                           
        # quick aside to increment subtotal with score
        subtotal += goal
                           
        #let's handle ties quickly
        if goal == 3:
            subtotal += SHAPE_SCORE[oppShape]
            print("my shape score: " + str(SHAPE_SCORE[oppShape]))
            print("subtotal: " + str(subtotal))
            total += subtotal
            print("total: " + str(total))
            continue
        
        # calculate shape based on oppShape and goal
        myShape = shapeFinder(oppShape, goal)
        print("myShape: " + myShape)
        subtotal += SHAPE_SCORE[myShape]
        print("myShape score: " + str(SHAPE_SCORE[myShape]) )
                           
        print("subtotal: " + str(subtotal))
        
        total += subtotal
        print("total: " + str(total))
        
        
    return total
    

    
if p1 is True:
    total_points = partOne()
else:
    total_points = partTwo()
    
print("final score: " + str(total_points))
```
